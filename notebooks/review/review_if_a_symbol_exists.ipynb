```python

import yfinance as yf
```

```python
symbol = "-P-HIZ"
```

```python
ticker = yf.Ticker(symbol)
try:
    price = ticker.fast_info.last_price
    name = ticker.info.get('longName', 'no name found')
    print('✅ Disponible')
    print('   last_price:', price)
    print('   longName:', name)
except Exception as e:
    print('❌ No disponible:', e)
```
